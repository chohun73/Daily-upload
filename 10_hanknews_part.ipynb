- 한국경제뉴스 기사 1개 땡겨와서 엑셀에 제목, 내용, 날짜 추출

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 뉴스 1개 크롤링
url = 'https://n.news.naver.com/mnews/article/015/0004993601'
html_doc = requests.get(url).text
html_doc
```

```python
soup = BeautifulSoup(html_doc,'html.parser')
# BeautifulSoup을 통해서 html을 보기 좋은 형태로 바꿈
soup
```

```python
# 기사 제목을 추출
title1 = soup.find('h2', 'media_end_head_headline')
title1.text
```

```python
# 기사 내용 추출
body1 = soup.find('article', 'go_trans _article_content')
body1.text
```

```python
# 딕셔너리 데이터 구조에 저희가 원하는 데이터를 담습니다.
data = {'뉴스url':[url], '제목':[title1.text], '내용':[body1.text]}

# 저희가 만든 데이터를 데이터프레임 구조로 만들어줍니다.
df = pd.DataFrame(data)

#CSV 파일로 저장합니다.
df.to_csv('news1_hank_index_False.csv', index = False)
df.to_csv('news1_hank_index_True.csv', index = True)
```

```python
url_list = []
title_list = []
body_list = []

url1 = 'https://n.news.naver.com/mnews/article/015/0004993601'
html_doc1 = requests.get(url1).text
soup1 = BeautifulSoup(html_doc1, 'html.parser')
title1 = soup1.find('h2', class_ = 'media_end_head_headline').text
body1 = soup1.find('article', class_= 'go_trans _article_content').text

url_list.append(url1)
title_list.append(title1)
body_list.append(body1)

url2 = 'https://n.news.naver.com/mnews/article/015/0004993586'
html_doc2 = requests.get(url2).text
soup2 = BeautifulSoup(html_doc2, 'html.parser')
title2 = soup2.find('h2', class_ = 'media_end_head_headline').text
body2 = soup2.find('article', class_= 'go_trans _article_content').text

url_list.append(url2)
title_list.append(title2)
body_list.append(body2)

data12 = {'뉴스url':url_list,'제목':title_list,'내용':body_list}

df12 = pd.DataFrame(data12)
df12.to_csv('news12_hank.csv', index = False)
```

```python
# 모듈화
url_list = []
title_list = []
body_list = []

urls = ['https://n.news.naver.com/mnews/article/015/0004993601',
        'https://n.news.naver.com/mnews/article/015/0004993586']

for url in urls:
    html_doc = requests.get(url).text
    soup = BeautifulSoup(html_doc,'html.parser')
    title = soup.find('h2', class_ = 'media_end_head_headline').text
    body = soup.find('article', class_= 'go_trans _article_content').text

    url_list.append(url)
    title_list.append(title)
    body_list.append(body)

data = {'뉴스url':url_list,'제목':title_list,'내용':body_list}

df = pd.DataFrame(data)
df.to_csv('news_hank_same.csv', index = False)
```
